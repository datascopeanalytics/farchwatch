```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_csv('./data/468104.csv')
```

```python
#df.columns
```

```python
#columns = [c for c in df.columns if c==c.upper() and c not in ['ELEVATION','LATITUDE','LONGITUDE']]
columns = [u'STATION', u'STATION_NAME', u'DATE', u'PRCP', u'SNWD', u'SNOW', u'TMAX', u'TMIN', u'TOBS', u'AWND']#, u'WSF2', u'WSF5', u'WT01', u'WT06', u'WT04', u'WT08', u'WT03']
```

```python
df = df.replace('-9999',np.NAN)
df = df.replace('9999',np.NAN)
df.DATE = pd.to_datetime(df.DATE.map(lambda x:str(x)),format="%Y%m%d")
```

```python
def faren(cel):
    if pd.isnull(cel):
        return cel
    else:
        return int(cel)*9/50+32
    
def day_of_year(dt):
    return int(datetime.strftime(dt,'%j'))
```

```python
df2 = df[columns]#[df.DATE>datetime(2010,1,1)]
df2['TMAX'] = df2.TMAX.map(faren)
df2['TMIN'] = df2.TMIN.map(faren)
df2['TOBS'] = df2.TOBS.map(faren)
df2['year'] = df2['DATE'].map(lambda x:x.year)
df2['month'] = df2['DATE'].map(lambda x:x.month)
df2['day'] = df2['DATE'].map(day_of_year)
df2=df2[df2.month < 5]
```

```python
df2.head()
```

```python
df3 = df2.groupby(['DATE']).mean()
```

```python
fig, ax = plt.subplots(figsize = (16,6))
plot_what = 'TOBS'

for year in range(1995,2000):
    slice = df3[df3.year==year]
    ax.plot(slice['day'],slice[plot_what],color='lightblue')

for year in range(2005,2010):
    slice = df3[df3.year==year]
    ax.plot(slice['day'],slice[plot_what],color='lightgreen')

    
slice = df3[df3.year==2014]
ax.plot(slice['day'],slice[plot_what],color='blue')
```
